# pyredshift from a notebook

`pyredshift.redshift.redshift(wave, flux)` works with plain numpy arrays.
The interactive window opens **outside** the notebook (a native window —
inline backends can't deliver mouse/key events); this cell blocks until
you quit with `q`, then the final annotated view is embedded below the
cell and the redshift returned.

Press `?` in the window (or click the `?` button) for the command
reference.

```python
import numpy as np
from astropy.io import fits

# One of the bundled example spectra: a z=2.5 SDSS quasar
with fits.open("Example-Spectra/SDSS-QSO-z2.50.fits") as h:
    tab = h["COADD"].data
    wave = 10.0 ** tab["loglam"].astype(float)
    flux = tab["flux"].astype(float)
wave.shape, flux.shape
```

```python
from pyredshift import redshift as redshift_module

z = redshift_module.redshift(wave, flux)   # blocks until you press q
z
```
